 1. My Data Contract

One row represents:
One content page for one client on one reporting date.

Tables used:
fact_content_daily_performance (March 2026)

Time window:
March 2026

Prediction / Ranking Goal:
Rank content pages that are most likely to need optimization based on historical engagement and search metrics.

Excluded Information:
Future information and any label-derived columns are excluded to avoid data leakage.

```python
con.execute(f"""
SELECT COUNT(*) AS total_rows
FROM read_parquet('{march_file}')
""").df()
```

```python
con.execute(f"""
SELECT
MIN(report_date) AS first_date,
MAX(report_date) AS last_date
FROM read_parquet('{march_file}')
""").df()
```

```python
con.execute(f"""
SELECT
COUNT(*) AS rows_available
FROM read_parquet('{march_file}')
WHERE gsc_data_available IS TRUE
""").df()
```

```python
features = con.execute(f"""
SELECT
content_hash_id,
gsc_impressions,
gsc_clicks,
gsc_avg_position,
ga4_pageviews,
ga4_engaged_sessions
FROM read_parquet('{march_file}')
LIMIT 10
""").df()
features
```

Feature Availability

gsc_impressions
Available immediately from Google Search Console.

gsc_clicks
Known before making any optimization decision.

 gsc_avg_position
Observed search ranking before optimization.

ga4_pageviews
Collected from Google Analytics before decision time.

ga4_engaged_sessions
Represents user engagement already observed before prediction.

```python
import pandas as pd

demo = features.copy()
demo["high_clicks"] = (demo["gsc_clicks"] > demo["gsc_clicks"].median()).astype(int)
demo["leaky_feature"] = demo["high_clicks"]

print(demo.head())
```

Leakage Experiment

I intentionally created `leaky_feature` directly from the target (`high_clicks`).

If this feature were used for training, the model would achieve unrealistically high performance because it already contains the answer.

After demonstrating the issue, I removed the leaky feature and kept only historical features available before the prediction time.

Limitation

This notebook uses only one month (March 2026) of warehouse data.
Seasonal trends and long-term behavior cannot be captured using a single month.
