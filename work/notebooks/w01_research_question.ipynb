I chose the Content Refresh Prediction lane.

The objective is to identify webpages that are likely declining in search performance and should be prioritized for content refresh. This lane helps content teams decide which pages to update first based on observed performance metrics.

2. Research Question

Research Question:

Which webpages should be prioritized for content refresh based on declining search performance?

Decision:
Identify pages that require content updates.

Action:
Editors refresh or improve those webpages.

Cost of Wrong Prediction
False Positive:
Refreshing a healthy page wastes time.

False Negative:
Ignoring a declining page may result in traffic loss.

```python
print(df["trend_direction"].value_counts())
print()
print("Average CTR:", round(df["ctr"].mean(),4))
print()
corr = df["search_volume"].corr(df["impressions_90d"])
print("Correlation:", round(corr,3))
```
