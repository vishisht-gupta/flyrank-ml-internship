This notebook frames my chosen machine learning lane for the FlyRank ML Internship. It explains the research question, the decision it supports, supporting evidence from the starter dataset, and the limits of the conclusions that can be drawn.

1. My Lane (or Freestyle) and Why

Content Refresh Prediction


I chose the Content Refresh Prediction lane because organizations often manage thousands of webpages, making it difficult to know which pages should be updated first. This project aims to identify pages that are likely declining in search performance so content teams can prioritize their work. During the starter notebooks, I observed that multiple signals such as impressions, click-through rate (CTR), and content age contribute to page performance, making this a suitable machine learning problem rather than relying on a single rule.

2. The Question: Decision, Action, Cost of a Wrong Call

Research Question

Which webpages should be prioritized for content refresh because they are likely declining in search performance?

Decision

The decision is to identify which webpages should be reviewed and updated first.

Action

A content editor or SEO team can refresh the selected webpages by improving their content, updating outdated information, optimizing keywords, or enhancing user experience.

Cost of a Wrong Recommendation

False Positive (predicting a page needs refresh when it doesn't):
- Time and resources are wasted updating a page that is already performing well.

False Negative (missing a page that actually needs refresh):
- A declining page continues to lose traffic, reducing visibility and potential business value.

```python
import pandas as pd

df = pd.read_csv("data/raw/content_refresh_anonymized.csv")

print("Dataset Shape")
print("Rows:", df.shape[0])
print("Columns:", df.shape[1])
print(df["trend_direction"].value_counts())

print("\nAverage CTR:", round(df["ctr"].mean(), 4))

corr = df["search_volume"].corr(df["impressions_90d"])
print("\nCorrelation between Search Volume and Impressions:", round(corr, 3))
```

```python
From the starter dataset:

- The dataset contains 30,000 webpages with 44 features describing search performance and content characteristics.
- 16,262 pages are labelled as declining (trend_direction = "down"), indicating that many pages may require attention.
- The average Click-Through Rate (CTR) across all pages is 0.5107.
- The correlation between search volume and impressions is 0.001, showing that keyword search volume alone is not a strong predictor of page impressions.

These observations suggest that multiple signals should be considered together, making this an appropriate machine learning problem rather than relying on a single rule.
```
