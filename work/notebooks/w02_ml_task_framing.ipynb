 ML-03 — Frame Your Lane as an ML Task

This notebook defines my chosen machine learning task for the FlyRank ML Internship. It explains the ML problem type, prediction target, evaluation metric, unit of analysis, and why machine learning is more suitable than a fixed rule.

1. My Lane as an ML Task

Chosen Lane

Content Refresh Prediction

Machine Learning Task

This is a binary classification problem.

The objective is to classify whether a webpage is likely to be declining in search performance or not.

Each webpage receives one of two labels:

- Declining
- Not Declining

The model predicts which pages should be prioritized for review and content refresh.

 2. Target or Proxy

 Target

The target variable is:

is_declining_label

This label is created from the dataset by checking whether:


trend_direction == "down"


Pages marked as down receive a value of 1, while all other pages receive 0.

This serves as the prediction target for the classification model.

3. Success Metric

The primary evaluation metric is Precision@50.

This metric measures how many of the top 50 webpages predicted as declining are actually declining.

Precision@50 is appropriate because content teams can only review a limited number of webpages at a time, so the highest-ranked recommendations should be as accurate as possible.

```python
import pandas as pd

df = pd.read_csv("data/raw/content_refresh_anonymized.csv")
df["is_declining_label"] = (
    df["trend_direction"].str.lower() == "down"
).astype(int)
print("Dataset Shape:", df.shape)
display(df.head())
print("\nTarget Distribution:")
print(df["is_declining_label"].value_counts())
```

4. The Unit of Analysis

The unit of analysis is one webpage.

Each row in the dataset represents a single webpage and contains information such as:

- Search volume
- Impressions
- CTR
- Average position
- Content age
- Word count
- Engagement metrics

The model learns from these features to predict whether a webpage is likely to be declining in search performance.

5. Why Machine Learning Beats a Fixed Rule

A fixed rule can only consider a few manually chosen conditions, such as "refresh pages older than 180 days with high impressions." While simple and easy to understand, such rules often miss complex interactions between multiple features.

Machine learning can combine signals such as impressions, CTR, average position, content age, engagement rate, and word count to identify patterns that are difficult to express with a single rule.

This allows the model to prioritize webpages more accurately and provide better decision support for content teams than a fixed rule alone.
